# Single layer perceptron

```python
x1 = 2
x2 = 5
x3 = 9
x4 = 1

w1 = 0.4
w2 = 0.6
w3 = 0.8
w4 = 1

#Bias: prevent from errors
b = 1

x_r1 , x_r2, x_r3, x_r4 = x1+w1, x2+w2 ,x3+w3, x4+w4
result = (x_r1 + x_r2 + x_r3 + x_r4) + b
print(f"result = {result}")

# activation function : decison on the basis of result
if result > 20:
  print("pass")
else:
  print("fail")
```

Multi layer perceptron

```python
#input Layer
x1 = 0.4
x2 = 0.3
x3 = 0.7
x4 = 0.6
print(f"Input layer \n input 1 = {x1}\n input 2 = {x2}\n input 3 = {x3}\n input 4 = {x4}")
#Hidden layer 1
# //////////////////////////////////////////////////////////////////////////
#weight of layer 1 perceptron 1 input 1
w1_1_1 = 0.05
w1_1_2 = 0.10
w1_1_3 = 0.15
w1_1_4 = 0.20
#hiden layer 1 perceptron 1
hl1_p1 = (x1*w1_1_1) + (x2*w1_1_2) + (x3*w1_1_3) + (x4*w1_1_4)
print(f"\nHidden layer 1\nhl1_p1 = {hl1_p1}")

#weight of layer 1 perceptron 2
w1_2_1 = 0.30
w1_2_2 = 0.35
w1_2_3 = 0.40
w1_2_4 = 0.45
hl1_p2 = (x1*w1_2_1) + (x2*w1_2_2) + (x3*w1_2_3) + (x4*w1_2_4)
print(f"hl1_p2 = {hl1_p2}")

#weight of layer 1 perceptron 3
w1_3_1 = 0.55
w1_3_2 = 0.60
w1_3_3 = 0.65
w1_3_4 = 0.70
hl1_p3 = (x1*w1_3_1) + (x2*w1_3_2) + (x3*w1_3_3) + (x4*w1_3_4)
print(f"hl1_p3 = {hl1_p3}")

#weight of layer 1 perceptron 4
w1_4_1 = 0.75
w1_4_2 = 0.80
w1_4_3 = 0.85
w1_4_4 = 0.90
hl1_p4 = (x1*w1_4_1) + (x2*w1_4_2) + (x3*w1_4_3) + (x4*w1_4_4)
print(f"hl1_p4 = {hl1_p4}")

#weight of layer 1 perceptron 5
w1_5_1 = 0.95
w1_5_2 = 1.00
w1_5_3 = 1.05
w1_5_4 = 1.10
hl1_p5 = (x1*w1_5_1) + (x2*w1_5_2) + (x3*w1_5_3) + (x4*w1_5_4)
print(f"hl1_p5 = {hl1_p5}")



#Hidden layer 2
# //////////////////////////////////////////////////////////////////////////
#weight of layer 2 perceptron 1 input 1
w2_1_1 = 1.15
w2_1_2 = 1.20
w2_1_3 = 1.25
w2_1_4 = 1.30
w2_1_5 = 1.35
#hiden layer 2 perceptron 1
hl2_p1 = (hl1_p1*w2_1_1) + (hl1_p2*w2_1_2) + (hl1_p3*w2_1_3) + (hl1_p4*w2_1_4) + (hl1_p5*w2_1_5)
print(f"\nHidden layer 2\nhl2_p1 = {hl2_p1}")

#weight of layer 2 perceptron 2
w2_2_1 = 1.40
w2_2_2 = 1.45
w2_2_3 = 1.50
w2_2_4 = 1.55
w2_2_5 = 1.60
hl2_p2 = (hl1_p1*w2_2_1) + (hl1_p2*w2_2_2) + (hl1_p3*w2_2_3) + (hl1_p4*w2_2_4) + (hl1_p5*w2_2_5)
print(f"hl2_p2 = {hl2_p2}")

#weight of layer 2 perceptron 3
w2_3_1 = 1.65
w2_3_2 = 1.70
w2_3_3 = 1.75
w2_3_4 = 1.80
w2_3_5 = 1.85
hl2_p3 = (hl1_p1*w2_3_1) + (hl1_p2*w2_3_2) + (hl1_p3*w2_3_3) + (hl1_p4*w2_3_4) + (hl1_p5*w2_3_5)
print(f"hl2_p3 = {hl2_p3}")

#weight of layer 2 perceptron 4
w2_4_1 = 1.90
w2_4_2 = 1.95
w2_4_3 = 2.00
w2_4_4 = 2.05
w2_4_5 = 2.10
hl2_p4 = (hl1_p1*w2_4_1) + (hl1_p2*w2_4_2) + (hl1_p3*w2_4_3) + (hl1_p4*w2_4_4) + (hl1_p5*w2_4_5)
print(f"hl2_p4 = {hl2_p4}")

#weight of layer 2 perceptron 5
w2_5_1 = 2.15
w2_5_2 = 2.20
w2_5_3 = 2.25
w2_5_4 = 2.30
w2_5_5 = 2.35
hl2_p5 = (hl1_p1*w2_5_1) + (hl1_p2*w2_5_2) + (hl1_p3*w2_5_3) + (hl1_p4*w2_5_4) + (hl1_p5*w2_5_5)
print(f"hl2_p5 = {hl2_p5}")



#Hidden layer 3
# //////////////////////////////////////////////////////////////////////////
#weight of layer 3 perceptron 1 input 1
w3_1_4 = 2.40
w3_1_1 = 2.45
w3_1_2 = 2.50
w3_1_5 = 2.65
w3_1_3 = 2.60
#hiden layer 3 perceptron 1
hl3_p1 = (hl2_p1*w2_1_1) + (hl2_p2*w2_1_2) + (hl2_p3*w2_1_3) + (hl2_p4*w2_1_4) + (hl2_p5*w2_1_5)
print(f"\nHidden layer 3\nhl3_p1 = {hl3_p1}")

#weight of layer 3 perceptron 2
w3_2_2 = 2.65
w3_2_1 = 2.70
w3_2_4 = 2.75
w3_2_3 = 2.80
w3_2_5 = 2.85
hl3_p2 = (hl2_p1*w2_2_1) + (hl2_p2*w2_2_2) + (hl2_p3*w2_2_3) + (hl2_p4*w2_2_4) + (hl2_p5*w2_2_5)
print(f"hl3_p2 = {hl2_p2}")

#weight of layer 3 perceptron 3
w3_3_1 = 2.90
w3_3_2 = 2.95
w3_3_3 = 2.00
w3_3_4 = 3.05
w3_3_5 = 3.10
hl3_p3 = (hl2_p1*w2_3_1) + (hl2_p2*w2_3_2) + (hl2_p3*w2_3_3) + (hl2_p4*w2_3_4) + (hl2_p5*w2_3_5)
print(f"hl3_p3 = {hl3_p3}")

#weight of layer 3 perceptron 4
w3_4_2 = 3.15
w3_4_1 = 3.20
w3_4_4 = 3.25
w3_4_3 = 3.30
w3_4_5 = 3.35
hl3_p4 = (hl2_p1*w2_4_1) + (hl2_p2*w2_4_2) + (hl2_p3*w2_4_3) + (hl2_p4*w2_4_4) + (hl2_p5*w2_4_5)
print(f"hl3_p4 = {hl3_p4}")

#weight of layer 3 perceptron 5
w3_5_3 = 3.40
w3_5_4 = 3.45
w3_5_1 = 3.50
w3_5_2 = 3.55
w3_5_5 = 3.60
hl3_p5 = (hl2_p1*w2_5_1) + (hl2_p2*w2_5_2) + (hl2_p3*w2_5_3) + (hl2_p4*w2_5_4) + (hl2_p5*w2_5_5)
print(f"hl3_p5 = {hl3_p5}")



#Hidden layer 4
# //////////////////////////////////////////////////////////////////////////
#weight of layer 4 perceptron 1 input 1
w4_1_1 = 3.65
w4_1_2 = 3.70
w4_1_3 = 3.75
w4_1_4 = 3.80
w4_1_5 = 3.85
#hiden layer 4 perceptron 1
hl4_p1 = (hl3_p1*w2_1_1) + (hl3_p2*w2_1_2) + (hl3_p3*w2_1_3) + (hl3_p4*w2_1_4) + (hl3_p5*w2_1_5)
print(f"\nHidden layer 4\nhl4_p1 = {hl4_p1}")

#weight of layer 4 perceptron 2
w4_2_1 = 4.95
w4_2_2 = 5.00
w4_2_3 = 5.10
w4_2_4 = 5.15
w4_2_5 = 5.20
hl4_p2 = (hl3_p1*w2_2_1) + (hl3_p2*w2_2_2) + (hl3_p3*w2_2_3) + (hl3_p4*w2_2_4) + (hl3_p5*w2_2_5)
print(f"hl4_p2 = {hl4_p2}")

#weight of layer 4 perceptron 3
w4_3_1 = 5.25
w4_3_2 = 5.30
w4_3_3 = 5.35
w4_3_4 = 5.40
w4_3_5 = 5.45
hl4_p3 = (hl3_p1*w2_3_1) + (hl3_p2*w2_3_2) + (hl3_p3*w2_3_3) + (hl3_p4*w2_3_4) + (hl3_p5*w2_3_5)
print(f"hl4_p3 = {hl4_p3}")

#weight of layer 4 perceptron 4
w4_4_1 = 5.50
w4_4_2 = 5.55
w4_4_3 = 5.60
w4_4_4 = 5.65
w4_4_5 = 5.70
hl4_p4 = (hl3_p1*w2_4_1) + (hl3_p2*w2_4_2) + (hl3_p3*w2_4_3) + (hl3_p4*w2_4_4) + (hl3_p5*w2_4_5)
print(f"hl4_p4 = {hl4_p4}")

#weight of layer 4 perceptron 5
w4_5_1 = 5.75
w4_5_2 = 5.80
w4_5_3 = 5.85
w4_5_4 = 5.95
w4_5_5 = 6.00
hl4_p5 = (hl3_p1*w2_5_1) + (hl3_p2*w2_5_2) + (hl3_p3*w2_5_3) + (hl3_p4*w2_5_4) + (hl3_p5*w2_5_5)
print(f"hl4_p5 = {hl4_p5}")



result = (hl4_p1 + hl4_p2 + hl4_p3 + hl4_p4 + hl4_p5)
print(f"\nResult Layer\nresult = {result}")

if result<5000 :
  print("output = 1")
else:
  print("output = 0")
```
